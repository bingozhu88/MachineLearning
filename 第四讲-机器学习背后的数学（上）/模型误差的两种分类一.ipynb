# 泛化误差的来源

问题内在的不确定性——》 噪声

数据永远不足——》 方差

模型本身的局限性——》 偏差

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures  #生成新特征， 1， x, x**2, x**3...，用这些特征做线性回归
import matplotlib.pyplot as plt
import numpy as np
```

产生数据的模型是
y = x + 0.1*x**2 + sin(x) + e, where e服从N（0,1）

```python
np.random.seed(1)
x = np.random.exponential(4, size = 100)
plt.figure(0)
plt.hist(x)
```

```python
y = np.polyval([0, 1, 0.1], x) + np.sin(x)
y_D = y + np.random.normal(0, 2, 100)

plt.figure()
plt.scatter(x, y, alpha= 0.5, marker='*')
plt.scatter(x, y_D, alpha= 0.3, c='r')
```

```python
def gen_data(N):
    x = np.random.exponential(4, size = N)
    #x = x.reshape(1000, 1)
    y = np.polyval([0, 1, 0.1], x)   + np.sin(x)
    #y = np.polyval(x, 0)+ np.sin(x)
    y_D = y + np.random.normal(0, 2, N)#.reshape(-1, 1)
    return x,y,y_D
    
x, y, yD = gen_data(1000)
print (yD.shape)
#print (y)
plt.scatter(x, y)
plt.scatter(x, yD, marker='*')
```

用于回归预测的模型是：y = ∑p WiX**i + e(误差)

```python
x = x.reshape(1000, 1)
poly = PolynomialFeatures(degree = 3)
x_ = poly.fit_transform(x)
plt.figure()
model = LinearRegression()
model.fit(x_, yD)
plt.scatter(x, model.predict(x_))
plt.scatter(x, yD)
```

# 2.1 生成测试点

```python
np.random.seed(0)
test = np.random.exponential(4, size =  1).reshape(1, 1)
y = np.polyval( [0, 1, 0.1], test) + np.sin(test)
print (y)
```

```python
def y_pred(test, N, degree = 1):
    #训练模型
    x, y, y_D = gen_data(N)
    x = x.reshape(N, 1)
    poly = PolynomialFeatures(degree = degree)
    x_ = poly.fit_transform(x)
    model = LinearRegression()
    model.fit(x_, y_D)
    #测试，预测
    test_ = poly.fit_transform(test)
    y_pred = model.predict(test_)
    return y_pred

y_pred(test, 1000)
```

# 2.2 结果分布图

```python
def trials(M, N, test, degree =2):
    y_preds = []
    y_Ds = []
    for i in range(M):
        y_preds.append(y_pred(test, N, degree= degree))
        y_D = np.polyval([0, 1, 0.1], test) + np.sin(test) + np.random.normal(0, 2)
        y_Ds.append(y_D)
    y_preds = np.array(y_preds).ravel()
    y_Ds = np.array(y_Ds).ravel()
    return y_preds, y_Ds

y_preds, y_Ds = trials(100, 1000, test)

plt.figure(0)
plt1 = plt.hist(y_preds)
plt.figure(1)
plt2 = plt.hist(y_Ds)

#print (y_preds)
```

```python
print (y_preds.mean(), y_preds.var(), y_Ds.mean(), y_Ds.var())
```

```python
#总误差为：
error = np.mean((y_preds - y_Ds)**2)
print (error)
```

```python
#方差为：
var = y_preds.var()
print(var)
```

```python
#偏差为：
bias = (y - y_preds.mean())**2
bias
```

总的误差=方差+偏差+噪声；方差和偏差之间存在平衡问题；训练集越多越大，方差越小；模型越简单（欠拟合），偏差越大；模型越复杂（过拟合），偏差越小
